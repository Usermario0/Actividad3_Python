```python
import re as re
unidades = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX']
decenas = ['X', 'XX', 'XXX', 'XL', 'L', 'LX', 'LXX', 'LXXX', 'XC']
centenas = ['C', 'CC', 'CCC', 'CD', 'D', 'DC', 'DCC', 'DCCC', 'CM']
millares = ['M', 'MM', 'MMM']
caracteresRomanos = ['I','V','X','L','C','D','M']

def decimal_a_romano(numero):
    if(numero < 10):
        return unidades[numero-1]
    
    if(numero >= 10 and numero < 100):
        cadena = str(numero)
        cadena = re.findall(r'\d', cadena)
        uno = int(cadena[1]) - 1
        if(uno == -1):
            return decenas[int(cadena[0]) - 1]
        return decenas[int(cadena[0]) - 1] + unidades[int(cadena[1]) - 1]
        
    if(numero >= 100 and numero < 1000):
        output = None
        cadena = str(numero)
        cadena = re.findall(r'\d', cadena)
        uno = int(cadena[2]) - 1
        diez = int(cadena[1]) - 1
        output = centenas[int(cadena[0]) - 1]
        if(diez != -1):
            output = output + decenas[diez]
        if(uno != -1):
            output = output + unidades[uno]
        return output
        
    if(numero >= 1000 and numero < 10000):
        output = None
        cadena = str(numero)
        cadena = re.findall(r'\d', cadena)
        uno = int(cadena[3]) - 1
        diez = int(cadena[2]) - 1
        cien = int(cadena[1]) - 1
        output = millares[int(cadena[0]) - 1]
        if (cien != -1):
            output = output + centenas[cien]
        if (diez != -1):
            output = output + decenas[diez]
        if (uno != -1):
            output = output + unidades[uno]    
        return output
        
        #return unidades[numero-1]

regexUnidades = "IV|VI*|IX|I+"
regexDecenas = "XL|LX*|XC|X+"
regexCentenas = "CD|DC*|CM|C+"
regexMillares = "M+"

# regexTotal = (regexMillares)*(regexCentenas)*(regexDecenas)*(regexUnidades)

def haremos_match(roman, regex, unidad):
    res = re.match(r"" + "^" + regex, roman)

    if res:
        romano_cortado = roman[res.end():]
        valor_obtenido = unidad.index(roman[:res.end()])+1
        return romano_cortado, valor_obtenido
    else:
        return roman, '0'

def romano_a_decimal(romano):
    decimal_total = False
    
    mil = haremos_match(romano, regexMillares, millares)
    romano = mil[0]
    decimal_total = str(mil[1])

    cien = haremos_match(romano, regexCentenas, centenas)
    romano = cien[0]
    decimal_total = decimal_total + str(cien[1])

    diez = haremos_match(romano, regexDecenas, decenas)
    romano = diez[0]
    decimal_total = decimal_total + str(diez[1])

    uno = haremos_match(romano, regexUnidades, unidades)
    decimal_total = decimal_total + str(uno[1])

    return int(decimal_total)

class Numero:
    def __init__(self, numero):
        self.romano = decimal_a_romano(numero)
        self.normal = numero

    def imprime(self):
        print(self.normal, self.romano)

    def suma_romano(self, otro_romano):
        romanoNuevoDecimal = romano_a_decimal(otro_romano)
        tempNormal = romanoNuevoDecimal + self.normal
        tempRomano = decimal_a_romano(romanoNuevoDecimal + self.normal)
        self.normal = tempNormal
        self.romano = tempRomano
    
    def is_romano(self, esRomano):
        cadenaSeparada = re.findall(r'\w',esRomano)
        for elemento in cadenaSeparada:
            if (caracteresRomanos.__contains__(elemento) == False):
                return False
        return True
        
# aux = Numero(1672)
# print(aux.is_romano('MDCXCII'))
# print(aux.is_romano('Hola'))
# lol = Numero(123)
# lol.imprime()
# lol.suma_romano('MMMXI')
```
